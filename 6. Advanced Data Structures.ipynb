## Exercice 1

Recently, we have contracted different streaming platform services. To sort all their content, we have created **3 stacks** to store what we’d like to see. The first contains documentaries, the second movies, and the third series. We have already seen some of the content of these stacks, so we have created **1 queue** with the contents already consumed.

Our goal is to generate **a stack** with the *playlist* of upcoming documentaries, movies and series that we haven't seen before but we want to see. To generate this stack we must take into account the following:

1. We want to alternate the contents according to their category in the following order: documentaries, films and series.
2. If we have already seen a particular content from a category, we will jump to the next category (e.g. if there is a series that we have already seen, we will add a documentary to the stack).
3. We don't want content in the same category consecutively (as long as we can avoid it)
4. We want to maintain the order of the contents as we have stored them in each of the stacks.

```python
# The following variables contain the documentaties, series and movies
# stacks, as well as the queue with watched contents.

documentaries_stack = ["Social Dilemma", "Wild Wild Country",
                       "Alpha Go", "Chasing Ice", "The Drawn Wall",
                       "Challenger: The Final Flight", "Fyre",
                       "Joan Didion: The Center Will not Hold", "Icaro",
                       "13th"]
series_stack = ["The Wire", "The Sopranos", "Chernobyl", "Big Little lies",
                "True Detective", "Breaking Bad", "Friends", "Black Mirror",
                "Peaky Blinders", "Mad Men"]
movies_stack = ["Dogville", "Persépolis", "Farenheit 451", "Take Shelter",
                "Carol", "Manhattan", "Seven", "The Ring", "Us",
                "Jurassik Park"]
watched_queue = ["Social Dilemma", "Chernobyl", "Take Shelter", "The Sopranos",
                 "Mad Men", "Friends", "Seven", "Fyre"]
```

```python
def print_stack(s, msg_bef=None, msg_after=None):
    """ Print a stack
    Args:
        s (:obj:`list`): list with stack elements
        msg_bef (:obj: `string`): message before the stack
        msg_after (:obj: `string`): message after the stack

    Returns:
        :obj:`string`: message with the string
    """

    if msg_bef:
        print(msg_bef)
    if len(s) == 0:
        max_len = 10
    else:
        max_len = max([len(e) for e in s]) + 4
    print("|" + " "*(max_len+2) + "|")
    for e in s[::-1]:
        print("| " + e + " "*(max_len-len(e)) + " |")
    print("|" + "_"*(max_len+2) + "|")
    print(msg_after+"\n" if msg_after else "\n")
```

```python
def check_queue(content, queue):
    """ Check if an element content is in the queue
    Args:
        content (:obj:`string`): list with stack elements
        queue (:obj: `list`): message before the stack

    Returns:
        watched (:obj:`boolean`): indicates whether the content
        is watched
    """

    queue_aux = queue[:]
    watched = False
    while len(queue_aux) > 0 and not watched:
        q = queue_aux.pop(0)
        if q == content:
            watched = True
    return watched
```

```python
result = []

total_len = (len(documentaries_stack) +
             len(series_stack) +
             len(movies_stack) -
             len(watched_queue))

while len(result) < total_len:
    if len(documentaries_stack) > 0:
        doc = documentaries_stack.pop()
        watched = check_queue(doc, watched_queue)
        if not watched:
            result.append(doc)
    if len(movies_stack) > 0:
        mov = movies_stack.pop()
        watched = check_queue(mov, watched_queue)
        if not watched:
            result.append(mov)
    if len(series_stack) > 0:
        ser = series_stack.pop()
        watched = check_queue(ser, watched_queue)
        if not watched:
            result.append(ser)

# Let's invert the result
ordered_result = []
while len(result) > 0:
    ordered_result.append(result.pop())

print_stack(ordered_result, "Final Stack:")
```

Create an ordered dictionary (`OrderedDict`) with the following fields:

- company_name
- country (you must extract the value of the `location` field)
- date (you must extract the value of the `datum` field)
- status_rocket
- rocket_cost
- status_mission

The dictionary must be indexed by the field **id** (eg. 0, 1, 2, ...) and sorted by the field
**date**. Also, only items with a valid `rocket_cost` value should be saved. The structure must be id0: id0_dict_with_its_values, id1: id0_dict_with_its_values, ...}.. For example, the `id0_dict_with_its_values` would be:


                (4020,
              {'company_name': 'US Air Force',
               'date': Timestamp('1964-09-01 15:00:00+0000', tz='UTC'),
               'status_rocket': 'StatusRetired',
               'detail': 'Titan IIIA | Transtage 1',
               'rocket_cost': 63.23,
               'country': 'USA',
               'status_mission': 'Failure'}),


```python
from collections import OrderedDict

df1 = df.drop('Unnamed: 0', axis=1)
df1 = df1.drop('Unnamed: 0.1', axis=1)

dates = []
countries = []

for i in list(range(0, len(df1))):
    try:
        date = df1.iloc[i, 2]
        strip = datetime.datetime.strptime(date, "%a %b %d, %Y %H:%M %Z")
        b1 = str(strip.strftime("%Y-%m-%d %H:%M:%S+%f"))
        b2 = date[-3:]
        final_date = "Timestamp('{}', tz='{}')".format(b1, b2)
        dates.append(final_date)
    
    except Exception:
        date = df1.iloc[i, 2]
        strip = datetime.datetime.strptime(date, "%a %b %d, %Y")
        b1 = str(strip.strftime("%Y-%m-%d %H:%M:%S+%f"))
        b2 = "UTC"
        final_date2 = "Timestamp('{}', tz='{}')".format(b1, b2)
        dates.append(final_date2)

    location = df1.iloc[i, 1]
    words = location.split(", ")
    countries.append(words[-1])

df1['country'] = countries
df1['data'] = dates
df2 = df1.sort_values('data')
df2 = df2.dropna(axis=0)

dict_rockets = OrderedDict()

for i in list(range(0, len(df2))):
    dict_rockets[df2.index[i]] = {
         'company_name': df2.iloc[i, 0],
         'date': df2.iloc[i, -1],
         'status_rocket': df2.iloc[i, -5],
         'detail': df2.iloc[i, 3],
         'rocket_cost': df2.iloc[i, -4],
         'country': df2.iloc[i, -2],
         'status_mission': df2.iloc[i, -3]
    }
dict_rockets
```

Define a **function** that receives the **dictionary from the previous exercise** as a parameter and:

The function must be prepared to work with very large dictionaries. Therefore, **it is necessary to avoid saving the individual values displayed on the screen in a list**.

Use the function with the dictionary from the previous exercise.

**Statement Correction**: Show only those missions located in the USA that meet the same statement requirements regardless of `company_name`.

```python
def sort_function(dictionary):
    """ Print information of dictionary
    Args:
        content (:obj:`OrderedDict`): dictionary with the elements to print

    Returns:
        :obj:`float`: average cost
        :obj:`float`: maximum cost
    """
    count_mission = 0
    max_cost = 0
    sum_cost = 0

    a = list(dictionary.items())

    for i in list(range(0, len(a))):
        row = list(a[i][1].values())
        if (row[2] == 'StatusRetired') & (row[5] == 'USA'):
            str_num = row[-3].strip()
            float_num = float(str_num.replace(",", ""))
            print(float_num)
            sum_cost += float_num
            count_mission += 1
            mitjana = sum_cost/count_mission

            if max_cost < float_num:
                max_cost = float_num

    return (mitjana, max_cost)


results = sort_function(dict_rockets)
print("The maximum cost is {} and the mean {}".format(results[1], results[0]))
```
