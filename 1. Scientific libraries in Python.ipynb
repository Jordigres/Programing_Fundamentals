### Exercice 1

Create and display the following matrices:

- A: 3x3 matrix with all the integers between 0 and 8.
- B: 3x3 identity matrix.
- C: 5x3 matrix with random real values.

Considering the above matrices, display on screen the following operations and explain the obtained results:

- Multiply A and B matrices.
- Multiply A and C matrices. 

Which methods did you chose? Why?

```python
# Load library
import numpy as np

# Matrix 1
A = np.arange(9).reshape(3, 3)

# Matrix 2
B = np.eye(3)

# Matrix 3
C = np.random.random((5,3))

print('Matrix A')
print(A)

print('\nMatrix B')
print(B)

print('\nMatrix C')
print(C)
```

```python
# Matrix multiplication of A and B
print("A.dot(B) result =\n{}".format(A.dot(B)))

# Element by element multiplication of A and B
print("\nA*B result =\n{}".format(A*B))

# Matrix multiplication of A and C
print("\nC.dot(A) result =\n{}".format(C.dot(A)))
```

**Answer** 

**A Section**

The matrix multiplication of two matrices A and B can be performed when the number of columns in matrix A is equal to the number of rows in matrix B. Also, it is possible to multiply element by element if the shape of both matrices is the same. 

Since A and B are both 3x3 matrices, they can be multiplied using both methods.  


**B Section**

Considering that A and C have a shape of 3x3 and 5x3 respectively, it is only possible to use the matrix multiplication.

### Exercice 3

Create a 7x7 matrix where each position `(i, j)` corresponds to `(i+1)^-j + 2` if `i` is a multiple of 3 and, otherwise, to `(i+1)^j - 3`.


```python
# Create empty matrix
m = np.empty((7, 7), dtype=int)

# Fill the matrix
for i in range(7):
    for j in range(7):
        if i % 3 == 0:
            m[i, j] = pow(1 + i, -j) + 2
        else:
            m[i, j] = pow(1 + i, j) - 3

print(m)
```

### Exercice 4

Normalize a random 5x5 matrix using its mean and standard deviation. Check if the mean and the standard deviation of the normalized matrix are close to 0 and 1 respectively. 


```python
# Create the random matrix
Z = np.random.random((5,5))

# Normalize
Z_norm = (Z - np.mean(Z)) / (np.std(Z))
print(Z_norm)

# Check if the new values follow a normal distribution
print('\nNormalized matrix mean: ',np.mean(Z_norm))
print('Normalized matrix standard deviation: ', np.std(Z_norm))
```

Plot on a single graphic the equations $\log_{10}(x)$, $\log_{2}(x)$ and $\ln(x)$ between the interval [0.1, 5] using a resolution of 0.02. The graphic must only show the `0.1, 5` range for the X axis and the `[-3, 3]` for the Y axis. 

Additionally, display $\log_{10}(x)$ using <span style="color: green">green</span>,  $\ln(x)$ using <span style="color: red">red</span> and $\log_{2}(x)$ using <span style="color: blue">blue</span>.

Plot a histogram showing the ratings. The plot must contain 20 green bins.

Show a scatter plot that displays the ratings against the price. In addition, differentiate each point considering the type of  book ([LabelEncoder](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html)) and show the legend.

**C section**

```python
# Load library
import pandas as pd

# Import dataframe
df = pd.read_csv("data/prog_book.csv")

# Plot histogram
n, bins, patches = plt.hist(df['Rating'], 20, density=1,
                            facecolor='g', alpha=0.75)

# Name of the axis
plt.xlabel('Rating')
plt.ylabel('Count')
```

### Exercice 5

Consider the 2018-2019 regular season of the NBA (no PlayOffs):

- Create a new column which stores information about the participation of each player in the PlayOffs (0 if the player wasn't in the PlayOffs and 1 if he was).

- Plot a histogram using the `height_cm` and `weight_kg` considering the value in the new column. Are the characteristics of the player different if they participated or not in the PlayOffs?

- Is there any relation between the height and the number of rebounds? Is there a difference between the players who participated in the PlayOffs and the ones who didn't?

```python
# Create new column full of 0
teams_PlayOff = np.unique(data_NBA_18_19_PlayOff['Team'])
data_NBA_18_19_Reg["PlayOff"] = 0

# Substitute some with 1
data_NBA_18_19_Reg.loc[data_NBA_18_19_Reg["Team"].isin(
    teams_PlayOff),'PlayOff'] = 1

# Create subsets
data_NBA_18_19_PO = data_NBA_18_19_Reg[data_NBA_18_19_Reg["PlayOff"]==1]
data_NBA_18_19_nPO = data_NBA_18_19_Reg[data_NBA_18_19_Reg["PlayOff"]==0]
```

```python
# Plot height and weight

# height
plt.hist(data_NBA_18_19_PO.height_cm, 12, density=True, 
         facecolor='g', alpha=0.75,stacked=True)
plt.hist(data_NBA_18_19_nPO.height_cm, 12, density=True, 
         facecolor='r', alpha=0.75,stacked=True)
plt.xlabel('Height (cm)')
plt.ylabel('Normalized count')
plt.title('Histogram of height distribution')
plt.legend(['PlayOffs players','Non PlayOffs players'])
plt.show()

# weight
plt.hist(data_NBA_18_19_PO.weight_kg, 12, density=True, 
         facecolor='g', alpha=0.75,stacked=True)
plt.hist(data_NBA_18_19_nPO.weight_kg, 12, density=True, 
         facecolor='r', alpha=0.75,stacked=True)
plt.xlabel('Weight (kg)')
plt.ylabel('Normalized count')
plt.title('Histogram of weight distribution')
plt.legend(['PlayOffs players','Non PlayOffs players'])
plt.show()
```

**Answer** 

Both histograms show very similar characteristics between both subgroups. Thus, it doesn't seem to have significant differences. 

```python
## PlayOff's players

# Plot number of rebounds against height
y = data_NBA_18_19_PO['REB']
x = data_NBA_18_19_PO['height_cm']
plt.plot(x, y, 'o')

# Calculate and add the regression line
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x+b)

# Add labels 
plt.xlabel('Height (cm)')
plt.ylabel('Rebounds')
plt.title('Height vs rebounds in Play Off players')
plt.show()


## Not-PlayOff's players

# Plot number of rebounds against height
y = data_NBA_18_19_nPO['REB']
x = data_NBA_18_19_nPO['height_cm']
plt.plot(x, y, 'o')

# Calculate and add the regression line
m, b = np.polyfit(x, y, 1)
plt.plot(x, m*x+b)

# Add labels 
plt.xlabel('Height (cm)')
plt.ylabel('Rebounds')
plt.title('Height vs rebounds in non Play Off players')
plt.show()
```

**Answer** 

Height and number of rebounds have a positive correlation. Thus, players who are tall tend to get more number of rebounds. This correlation exists on both subgroups of players, however, it doesn't seem to be a significant difference between the strenght of these correlations.

### Exercice 1

Python is a programming language widely used to process images. In this exercice, we will use a racoon image and its black and white (BW) version. 

- Load the images `racoon.png` and `racoonBW.png` in the `data`. Print the size of each image. Why do they have different sizes?

- Apply to the BW image a gaussian filter using a kernel size of 3 and 5. Then, apply a [median filter](https://docs.scipy.org/doc/scipy/reference/generated/scipy.ndimage.median_filter.html). Explain the different filter effects.

```python
# Load libraries
import matplotlib.image as mpimg
from scipy import ndimage

# Load images
racoon = mpimg.imread('data/racoon.png')
racoonBW = mpimg.imread('data/racoon_BW.png')

# Display both images in a figure
fig, axs = plt.subplots(1,2,figsize=(15, 5))
axs[0].imshow(racoon)
axs[1].imshow(racoonBW, cmap=plt.cm.gray)

print("Color image size: ",racoon.shape)
print("BW image size: ",racoonBW.shape)
```

**Answer** 

The colored image has a third dimension that represents the information of the primary colors (RGB). The first plane contains the red of every pixel in the image, the second the green, and the third the blue. 

Since the BW image doesn't have color, it only has 2 dimensions. 

```python
# Filters
blurred_face = ndimage.gaussian_filter(racoonBW, sigma=3)
very_blurred = ndimage.gaussian_filter(racoonBW, sigma=5)
med_denoised = ndimage.median_filter(racoonBW, 3)

# Display filtered images
fig, axs = plt.subplots(1,3,figsize=(15, 5))
axs[0].imshow(blurred_face, cmap=plt.cm.gray)  
axs[1].imshow(very_blurred, cmap=plt.cm.gray)  
axs[2].imshow(med_denoised, cmap=plt.cm.gray)  
```

**Answer** 

The gaussian filter is a linear filter usually used to blur the image or to reduce the noise. This type of filter only blurs the edges and reduces the contrast. 

The median filter is a non-linear filter usually used to reduce the noise of the image while maintaining the edges relatively sharp.

As can be observed in the images, if the gaussian filter uses a bigger kernel size, the blur effect is intensified, and the contrast is reduced. Instead, while the mean filter blurs the image, it still mantains the edges.

In this exercice, we want to compare the NBA regular season with the ACB in the 2018-2019 season. Generate a bar plot for each league containing the number of players according to their country. 

Note:

- Plot both leagues in different graphics.

```python
# Select ACB and season 18-19
data_ACB_18_19=data[(data["League"] == 'Spanish-ACB') & (data["Season"] == '2018 - 2019')]

# Obtain nationalities
nat_ACB = list(data_ACB_18_19['nationality'])
nat_NBA_Reg = list(data_NBA_18_19[data_NBA_18_19['Stage']=="Regular_Season"]['nationality'])

# Obtain NBA first nationalities
nat_NBA_Reg_clean=[]
for i in nat_NBA_Reg:
    nat_NBA_Reg_clean.append(i.split(' /')[0])
    
# Obtain ACB first nationalities
nat_ACB_clean=[]
for i in nat_ACB:
    nat_ACB_clean.append(i.split(' /')[0])
```

```python
# Add first nationalities into the dataframes
df_nat_NBA_Reg_clean=pd.DataFrame(nat_NBA_Reg_clean, columns=["nationality"])
df_nat_ACB_clean=pd.DataFrame(nat_ACB_clean, columns=["nationality"])

# Group by nationalities
# NBA
df_nat_NBA_Reg_clean_grouped = df_nat_NBA_Reg_clean.groupby("nationality")
df_nat_NBA_Reg_clean_count = df_nat_NBA_Reg_clean_grouped['nationality'].count()

# ACB
df_nat_ACB_clean_grouped = df_nat_ACB_clean.groupby("nationality")
df_nat_ACB_clean_count = df_nat_ACB_clean_grouped['nationality'].count()
```
